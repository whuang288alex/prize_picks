# Player Level Stats

```python
from nba_api.stats.endpoints import playergamelog
from nba_api.stats.static import players
from nba_api.stats.endpoints import boxscoreadvancedv2

p = players.find_players_by_full_name('james harden')[0]
game_log = playergamelog.PlayerGameLog(player_id=p['id'])
data_dict = game_log.get_dict()
```

```python
for game_results in data_dict["resultSets"]:
    if len(game_results['rowSet']) == 0:
        print('No games found for this player.')
    else:
        print('Found {} games.'.format(len(game_results['rowSet'])))
        for this_game_result in game_results['rowSet']:
            for key, value in zip(game_results['headers'], this_game_result):
                print(key, value)
            
            try:
                game_id = this_game_result[game_results['headers'].index('Game_ID')]
            except:
                game_id = this_game_result[game_results['headers'].index('GAME_ID')]
            game_log = boxscoreadvancedv2.BoxScoreAdvancedV2(game_id=game_id)
            data_dict = game_log.get_dict()
            player_stats, team_stats = data_dict["resultSets"]
            print("\n\n")
    break
```

```python
for game_results in data_dict["resultSets"]:
    if len(game_results['rowSet']) == 0:
        print('No games found for this player.')
    else:
        print('Found {} games.'.format(len(game_results['rowSet'])))
        for this_game_result in game_results['rowSet']:
            for key, value in zip(game_results['headers'], this_game_result):
                print(key, value)
            break
    break
```
